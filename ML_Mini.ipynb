```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Parkinsson disease.csv")
data
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['name'] = label_encoder.fit_transform(data['name'])
```

```python
data.isnull().sum()
```

```python
data.describe()
```

```python
data['status'].value_counts()
```

```python
X=data.drop(columns = ['status'], axis=1)
y=data['status']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
r=RandomForestClassifier()
r.fit(X_train,y_train)
prediction3=r.predict(X_test)
print("****ACCURACY BY USING RANDOM FOREST CLASSFIER ALGORITHM****:\n")
print(" ACCURACY = ",accuracy_score(y_test,prediction3))
```

```python
from sklearn.metrics import confusion_matrix
print("****CONFUSION MATRIX FOR RANDOM FOREST ALGORITHM****\n")
confusion_matrix(y_test,prediction3)
```

```python
from sklearn.metrics import classification_report
print("*****CLASSIFICATION REPORT FOR RANDOM FOREST ALGORITHM*****\n")
print(classification_report(y_test,prediction3))
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Setting the aesthetics for the plots
sns.set(style='whitegrid')
```

```python

# Histograms for numerical features
data.hist(bins=15, figsize=(22, 15), edgecolor='black')
plt.suptitle('Histograms of Numerical Features', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20, 15))
corr = data.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```

```python




# Scatter plot for a pair of features (example: MDVP:Fo(Hz) vs MDVP:Fhi(Hz))
plt.figure(figsize=(8, 6))
sns.scatterplot(x='MDVP:Fo(Hz)', y='MDVP:Fhi(Hz)', hue='status', data=data, palette='coolwarm', s=100)
plt.title('Scatter Plot of MDVP:Fo(Hz) vs MDVP:Fhi(Hz)', fontsize=16)
plt.xlabel('MDVP:Fo(Hz)')
plt.ylabel('MDVP:Fhi(Hz)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=data, x='MDVP:Fo(Hz)', hue='status', fill=True, common_norm=False, palette='coolwarm', alpha=0.5)
plt.title('Density Plot of MDVP:Fo(Hz) by Status', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=data.select_dtypes(include=['float64', 'int64']))
plt.title('Boxplots of Numerical Features', fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='status', y='MDVP:Fhi(Hz)', data=data, palette='coolwarm')
plt.title('Swarm Plot of MDVP:Fhi(Hz) by Status', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='status', y='MDVP:Fo(Hz)', data=data, palette='coolwarm')
plt.title('Violin Plot of MDVP:Fo(Hz) by Status', fontsize=16)
plt.show()
```
